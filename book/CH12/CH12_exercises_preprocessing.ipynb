```python
df = df[16150:]
```

```python
df.loc[:,'datetime'] = pd.to_datetime(df[['year','month','day','hour']])
df.head()
```

```python
df = df.drop(['year', 'month', 'day', 'hour'], axis=1)
df.head()
```

```python
hour = df.datetime.dt.hour
num_cols = df.select_dtypes(include='number').columns
df[num_cols] = df.groupby(hour)[num_cols].transform(lambda s: s.fillna(s.mean()))
df[num_cols] = df[num_cols].interpolate(method='linear').bfill().ffill()
```

```python
fig, ax = plt.subplots(figsize=(13,6))

ax.plot(df['CO'])
ax.set_xlabel('Date')

fig.autofmt_xdate()
plt.tight_layout()
```

```python
df['wd'] = df['wd'].fillna(df['wd'].value_counts().index.tolist()[0])
```
